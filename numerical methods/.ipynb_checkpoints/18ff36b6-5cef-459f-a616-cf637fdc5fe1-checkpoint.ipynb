# Описание данных

### Признаки    

- `DateCrawled` — дата скачивания анкеты из базы
- `VehicleType` — тип автомобильного кузова
- `RegistrationYear` — год регистрации автомобиля
- `Gearbox` — тип коробки передач
- `Power` — мощность (л. с.)
- `Model` — модель автомобиля
- `Kilometer` — пробег (км)
- `RegistrationMonth` — месяц регистрации автомобиля
- `FuelType` — тип топлива
- `Brand` — марка автомобиля
- `Repaired` — была машина в ремонте или нет
- `DateCreated` — дата создания анкеты
- `NumberOfPictures` — количество фотографий автомобиля
- `PostalCode` — почтовый индекс владельца анкеты (пользователя)
- `LastSeen` — дата последней активности пользователя   

### Целевой признак   

- `Price` — цена (евро)

## Обучение моделей

```python
# добавим функцию RMSE
def rmse(true, pred):
    return sqrt(mean_squared_error(true, pred))

rmse_scorer = make_scorer(rmse, greater_is_better=False)
```

```python
# разделим таблицу на признаки и целевой признак
features = df.drop(['price'], axis=1)
target = df['price']
```

```python
# разделим на тестовую и обучающую выборку данных
(features_train,
 features_test,
 target_train,
 target_test) = train_test_split(features,
                                    target,
                                    test_size=.35,
                                    random_state=12345)
```

### Линейная модель

```python
features_train_lr = features_train.copy()
features_test_lr = features_test.copy()
```

```python
# создадим список столбцов с распределением категориальных признаков
ohe_features_lr_obj = features_train.select_dtypes(include='object').columns.to_list()
ohe_features_lr = ohe_features_lr_obj + ['kilometer','registration_year']
ohe_features_lr
```

```python
encoder_ohe = OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False)
encoder_ohe.fit(features_train_lr[ohe_features_lr])

# прменим OneHotEncoder на обучающей выборке
features_train_lr[
    encoder_ohe.get_feature_names_out()
                    ] = encoder_ohe.transform(features_train_lr[ohe_features_lr])
features_train_lr = features_train_lr.drop(ohe_features_lr, axis=1)


# на тестовой выборке
features_test_lr[
    encoder_ohe.get_feature_names_out()
                    ] = encoder_ohe.transform(features_test_lr[ohe_features_lr])

features_test_lr = features_test_lr.drop(ohe_features_lr, axis=1)
```

```python
# масстабируем численные признаки
numbers = ['power']
```

```python
features_train_scaled = features_train_lr.copy()
features_test_scaled = features_test_lr.copy()
```

```python
scaler = StandardScaler()
scaler.fit(features_train_lr[numbers])
pd.options.mode.chained_assignment = None

features_train_scaled[numbers] = scaler.transform(features_train_scaled[numbers])
features_test_scaled[numbers] = scaler.transform(features_test_scaled[numbers])
```

```python
%%time
alphas = np.array([1,0.1,0.001,0])
param_grid = dict(alpha = alphas)

model_lr = Ridge(random_state=12345)
grid_lr = GridSearchCV(estimator = model_lr,
                       param_grid = param_grid,
                       cv=5,
                       scoring='neg_root_mean_squared_error',
                       n_jobs=-1,
                       error_score='raise')
grid_lr.fit(features_train_scaled, target_train)
rmse_lr = -grid_lr.best_score_
model_lr = grid_lr.best_estimator_

fit_time_start = time.time()
model_lr.fit(features_train_scaled, target_train)
fit_time_end = time.time()
pred = model_lr.predict(features_train_scaled)
predict_time_end = time.time()

# время обучения и предсказания модели сохраним
lr_fit_time = fit_time_end - fit_time_start
lr_pred_time = predict_time_end - fit_time_end

print('RMSE работы линейной модели на кросс-валидации:', rmse_lr)
print('Время обучения модели:', lr_fit_time, 'сек')
print('Время предсказания модели:', lr_pred_time, 'сек')
```

### CatBoostRegressor

```python
# выделим столбцы с категорильными данными
cat_features = ['vehicle_type', 'gearbox', 'model', 'fuel_type', 'brand', 'repaired']
```

```python
%%time
model_cb = CatBoostRegressor(loss_function='RMSE', random_state=1, verbose=0, cat_features=cat_features)
grid_cb_ = {'iterations': [10, 50, 100],
        'learning_rate': [0.03, 0.1],
        'depth': [4, 6, 10]}
grid_cb = GridSearchCV(model_cb, param_grid=grid_cb_, cv=5, scoring='neg_root_mean_squared_error')
grid_cb.fit(features_train, target_train)
rmse_cb = -grid_cb.best_score_
model_cb = grid_cb.best_estimator_

fit_time_start = time.time()
model_cb.fit(features_train, target_train)
fit_time_end = time.time()
pred = model_cb.predict(features_train)
predict_time_end = time.time()

# время обучения и предсказания модели сохраним
cb_fit_time = fit_time_end - fit_time_start
cb_pred_time = predict_time_end - fit_time_end

print('RMSE работы CatBoost на тестовой выборке:', rmse_cb)
print('Время обучения модели:', cb_fit_time, 'сек')
print('Время предсказания модели:', cb_pred_time, 'сек')
```

### LightGBM

```python
# преобразеум тип данных категориальных признаков
features_train[cat_features] = features_train[cat_features].astype('category')
features_test[cat_features] = features_test[cat_features].astype('category')
```

```python
%%time
lgb_model = lgb.LGBMRegressor(random_state=1)
parametrs = {'n_estimators': [10, 50, 100],
             'learning_rate': [0.03, 0.1],
             'max_depth': [1, 10, 15]}
grid_lgb = GridSearchCV(lgb_model, param_grid=parametrs, cv=5, scoring=rmse_scorer)
grid_lgb.fit(features_train, target_train)
rmse_lgb = -grid_lgb.best_score_
model_lgb = grid_lgb.best_estimator_


fit_time_start = time.time()
model_lgb.fit(features_train, target_train)
fit_time_end = time.time()
pred = model_lgb.predict(features_train)
predict_time_end = time.time()

# время обучения и предсказания модели сохраним
lgb_fit_time = fit_time_end - fit_time_start
lgb_pred_time = predict_time_end - fit_time_end

print('RMSE работы модели LightGBM на тестовой выборке:', rmse_lgb)
print('Время обучения модели:', lgb_fit_time, 'сек')
print('Время предсказания модели:', lgb_pred_time, 'сек')
```

## Анализ моделей

```python
#рассмотри полученные результаты
results = pd.DataFrame({'Models':['LR', 'CB', 'LGBM'],
                       'rmse': [rmse_lr, rmse_cb, rmse_lgb],
                       'train_time, sec': [lr_fit_time, cb_fit_time, lgb_fit_time],
                       'work_time, sec': [lr_pred_time, cb_pred_time, lgb_pred_time]})
```
